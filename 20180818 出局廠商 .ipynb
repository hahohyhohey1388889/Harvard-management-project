```python
def result(file18, file19):
    result1901 = []
    for i in range(97):
        a = list(file18['18調查品項'].unique())[i]##行動電話系統業者、數據網路服務...
        data18 = file18[file18['18調查品項'] == a ].reset_index()## 18調查品項 == 行動電話系統業者...的紫dataframe
        data19 = file19[file19['19調查品項'] == a].reset_index()## 19調查品項 == 行動電話系統業者...的紫dataframe
        #data18['18廠商名稱'][0] not in list(data19['19廠商名稱'][0:3])
        for j in range(3):
            if data18['18廠商名稱'][j] not in list(data19['19廠商名稱'][0:3]):
                if data18['18廠商名稱'][j] in data19['19廠商名稱'].tolist():##條件
                    bb = data18['18廠商名稱'][j]
                    result1901.append(data18['18調查品項'][0])
                    result1901.append(bb)
                    result1901.append(data18['18廠商排名'][j])
                    result1901.append(int(data19[data19['19廠商名稱']==bb].iloc[:,3]))
                    if int(data19[data19['19廠商名稱']==bb].iloc[:,3]) <=5:
                        result1901.append('否')
                    else:
                        result1901.append('是')
        if data18.iloc[3,3] == 3:
            if data18['18廠商名稱'][3] not in list(data19['19廠商名稱'][0:3]):
                if data18['18廠商名稱'][3] in data19['19廠商名稱'].tolist():##條件
                    bb = data18['18廠商名稱'][3]
                    result1901.append(data18['18調查品項'][0])
                    result1901.append(bb)
                    result1901.append(data18['18廠商排名'][3])
                    result1901.append(int(data19[data19['19廠商名稱']==bb].iloc[:,3]))
                    if int(data19[data19['19廠商名稱']==bb].iloc[:,3]) <=5:
                        result1901.append('否')
                    else:
                        result1901.append('是')
        
    return result1901
```
